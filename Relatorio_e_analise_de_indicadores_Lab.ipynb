## Tubos por mes

```python
#Agrupando pelo local de coleta e ano/mes do cadastro para calcular o numero de Tubos utilizados
local_meses_tubo_exames = situação_meses.groupby(['Local','ano_mes_cadastro','Tubo']).agg(
    N_Tubos=('O.S.','nunique') # Nesse calculo conseguimos achar quantos tubos cada um dos pacientes usaram em cada OS
    ).reset_index()

local_meses_tubo_exames.tail(2)
```

```python
#Pivotando a tabela
local_meses_tubo = pd.pivot_table(local_meses_tubo_exames, values='N_Tubos', index=['Local','Tubo'], columns='ano_mes_cadastro', fill_value=0)

local_meses_tubo.tail(2)
```

```python
# Converter nomes de coluna em datetime
local_meses_tubo.columns = pd.to_datetime(local_meses_tubo.columns, format='%Y/%m/%d')

# Ordenar colunas em ordem decrescente
local_meses_tubo = local_meses_tubo.sort_index(axis=1, ascending=False)

# Converter nomes de coluna de datetime para string - esse passo é importante para o momento da exportação do arquivo
# Pois ao tentar exportar para o google sheets em formato datatime ele da erro
local_meses_tubo.columns = local_meses_tubo.columns.map(lambda x: x.strftime('%Y/%m/%d'))

#Total
local_meses_tubo['Total'] = local_meses_tubo.sum(axis=1)

#Pasando total para frente
cols = local_meses_tubo.columns.tolist()
cols = cols[-1:] + cols[:-1]
local_meses_tubo = local_meses_tubo[cols]
```

# Analise de tendencias

```python
# Função para encontrar os exames que aparecem junto com o exame selecionado
def encontrar_exames_similares(Cod_Exame, unidade):
    # Filtrar o dataframe pelas criando um DF com as OS que tenham o exame e unidade selecionanda.
    filtro_os = (situação_meses['Cod. Exame'] == Cod_Exame) & (situação_meses['Local'] == unidade)
    os_filtradas = situação_meses.loc[filtro_os, 'O.S.'].unique()

    # Filtrar o dataframe pelos exames com as O.S. filtradas
    filtro_exames = situação_meses['O.S.'].isin(os_filtradas)
    df_filtrado = situação_meses.loc[filtro_exames]

    # Filtrar os exames que aparecem nas mesmas O.S. do exame selecionado
    exames_frequencia = df_filtrado.groupby('O.S.')['Exame'].apply(list)
    exames_frequencia = exames_frequencia.explode().value_counts()
    exames_frequencia = exames_frequencia[exames_frequencia.index != Cod_Exame]

    return exames_frequencia.head(20)

# Função interativa para selecionar o Alfa Exame e a Unidade
@interact
def selecionar_exame(Cod_Exame=situação_meses['Cod. Exame'].unique(), unidade=situação_meses['Local'].unique()):
    exames_similares = encontrar_exames_similares(Cod_Exame, unidade)
    print('Os exames com correlação são :')
    print(exames_similares)

    # Filtrar o dataframe pelos exames selecionados
    filtro_exames_selecionados = situação_meses['Exame'].isin(exames_similares.index)
    df_exames_selecionados = situação_meses.loc[filtro_exames_selecionados]

    # Agrupar os exames por data e contar a frequência
    df_exames_frequencia = df_exames_selecionados.groupby(['Data Cad', 'Exame']).size().unstack()
```
